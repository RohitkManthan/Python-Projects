# Does Rating Affect The Revenue?

```python
plt.bar(df['Rating'],df['Revenue'])
plt.xlabel('Rating')
plt.ylabel('Revenue')
plt.title('Rating VS Revenue')
plt.show()
print("Interpretation:\nBased on these observations, we can infer that higher ratings are likely associated with higher revenue.\nThis makes sense intuitively - products or services with better ratings tend to be more popular and in demand, leading to increased sales and revenue.")
```

# Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
def classify_rating(Rating):
    if Rating >= 8.5:
        return 'Excellent'
    elif 7 <= Rating < 8.5:
        return 'Good'
    else:
        return 'Average'

df['rating_category'] = df['Rating'].apply(classify_rating)

df
```
